### Deal with Dataframe (tidy up)

```python
import pandas as pd
```

```python
df = pd.read_excel("assignment-01-data-unformated.xlsx", header=[0, 1], skipinitialspace=True)
df
```

```python
df_drop_loc = df.drop(['Unnamed: 0_level_0', 'Unnamed: 1_level_0'], axis=1)
df_drop_loc
```

```python
new_df = df_drop_loc.stack()
new_df
```

```python
new_df.reset_index(level=[0,1], inplace=True)
```

```python
final_df = new_df.rename(index=str, columns={"level_0":"site", "name":"year"})
final_df
```

```python
final_df.to_excel("result1.xlsx")
```
